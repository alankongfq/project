```python
import numpy as np
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pk

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

### Scrap only Australia

```python
Countries = {"AU":"Australia"}

Target_cities= {'AU':["Sydney","Melbourne","Brisbane","Perth", "Adelaide", "Canberra", "Darwin", "Hobart"]}

URL ={"AU":"https://au.indeed.com/jobs"}
```

```python
## maximum search results
max_results_per_city = 1000
## Search parameters on url
## placeholders
parameters = {'q': 'data scientist', 'radius': '100', 'start':1}
jobs = ['data scientist','data analyst', 'data engineer', 'big data', 'cloud', 'business intelligence']
```

```python
def scrape_page_to_df(url, url_params, country):

    # create a empty dictionary to store extracted information for each page
    scraped_data = {'location': [],
                  'company': [],
                  'title': [],
                  'salary': [],
                  'description': [],
                  'review': [],
                  'star': [],
                  'country': []
                  }

    html = requests.get(url, params=url_params)

    # make sure the response status is ok
    assert html.status_code == requests.codes.ok
    
    soup = BeautifulSoup(html.text, 'lxml')

    # function to extract results for each search page

    def extract_results(soup):
        return soup.find_all('div', class_='result')

    results = extract_results(soup)

    ## input html for each search object and extract jobs information

    def extract_location(result):
        """extract job location"""
        try:
            location = result.find('span', class_='location').get_text().strip()
            return location
        except:
            return None
    
    def extract_company(result):
        """extract the name of the company"""
        try:
            company = result.find('span', class_='company').get_text().strip()
            return company
        except:
            return None

    def extract_title(result):
        """extract the job title"""
        try:
            title = result.find('a', attrs={'data-tn-element': "jobTitle"}).get('title')
            return title
        except:
            return None


    def extract_salary(result):
        """extract the salary"""
        try:
            salary = result.find('td', class_='snip').\
            find('span', class_='no-wrap').\
            get_text().strip()
            return salary
        except:
            return None


    def extract_description(result):
        """extract job description snippet"""
        try:
            description = result.find('span', class_='summary').get_text().strip()
            return description
        except:
            return None


    def extract_review(result):
        """extract the number of reviews for the company"""
        try:
            review = result.find('a', attrs={'data-tn-element': "reviewStars"})
            review = review.find('span', class_="slNoUnderline")
            review = review.get_text().strip()
            # extract only the number
            review = review.replace(',', '').replace(' reviews', '')
            return review
        except:
            return None            

    
    def extract_star(result):
        """extract a number (width) that is proportional to the number of stars
        shown for the company"""
        try:
            # the 'style' attribute dictates how many stars are filled with color
            star = result.find('span', class_='rating').get('style')
            # extract only the number
            star = star.replace('width:', '').replace('px', '')
            return star
        except:
            return None


    # append extracted info to the correspond list
    for result in results:
        scraped_data['location'].append(extract_location(result))
        scraped_data['company'].append(extract_company(result))
        scraped_data['title'].append(extract_title(result))
        scraped_data['salary'].append(extract_salary(result))
        scraped_data['description'].append(extract_description(result))
        scraped_data['review'].append(extract_review(result))
        scraped_data['star'].append(extract_star(result))
        scraped_data['country'].append(country)

    # convert the dictionary to a pandas dataframe and returns it
    return pd.DataFrame(scraped_data)
```

```python
def remove_duplicates(df):
    """remove duplicates and returns a new df"""
    
    nrows_before = df.shape[0]
    df.drop_duplicates(subset=['company', 'country','description',
                               'location', 'salary', 'title'],
                       keep='last', inplace=True)
    nrows_after = df.shape[0]
    
    print('{} rows remain after removing duplicates from {} rows.'.format(
        nrows_after, nrows_before))
    print('{} rows have salary info; {} rows have yearly salary info.'.format(
      df.salary.notnull().sum(), df.salary.str.contains('year').sum()))
    return df
```

```python
def scrapper(CountryCode):
    print('Current system time: {}'.format(time.ctime()))
 
    # scrape data and save to dataframe
    start_time = time.time()
    
    #Retrieve Parameters to scrape with based on input of Country Code
    url = URL[CountryCode]
    locations = Target_cities[CountryCode]
    country = Countries[CountryCode]
    
    #Create an empty place holder df, search through every location in that country, but only 1 results, just to get the title and columns
    df = scrape_page_to_df(url,parameters,country)
    
    for loc in locations:
        for work in jobs:
            for start in range(0, max_results_per_city, 10):
            
              
                url_params = parameters.copy()
                #update the job with the target job that we want, city for target city that we are looking for and start refers to the current page number being scrapped
                url_params.update({'l': loc,'q': work, 'start': start})


                #insert code to put the scrap stuff into a df here, after each round of loop, concat into a df
                df = pd.concat([df,scrape_page_to_df(url, url_params,country)],axis=0)
              
        print('Finished scraping {}'.format(loc))
    total_time = (time.time() - start_time) / 60
    print('Scraping run time: {:.1f} minutes'.format(total_time))

    
    
    
    # remove duplicates
    df = remove_duplicates(df)
    print('Script finished at {}\n'.format(time.ctime()))
    
    #returns the final df
    return df
```

```python
### seem that there is some fixed format for salary
### split salary by ' ' and get the length of the list after split
### Also, the last word in the list represent a duration corresponding to the pay.
### Will have to convert everything into a uniform time scale --- choosing hour as its the smallest
```

```python
## form 2 columns, 1 columns for len of salary list and 1 columns for the duration of the list (at the last word of the list)
```

```python
salary_len = []
salary_duration = []
for i in range(0, df2.shape[0], 1):
    k = df2.salary[i].split(' ')
    
    salary_duration.append(k[-1])
    salary_len.append(len(k))  
    
df2['salary_len'] = salary_len
df2['salary_duration'] = salary_duration
```

```python
df2.salary_len.value_counts()
```

```python
## so the len of the list is either 3 or 5.
## format of len 5 ---- [salarymin, - , salarymax, 'a', duration]
## format of len 3 ---- [salary, 'a', duration]

## therefore 
## for len 5, we can get the average between the salary range min and max.
## for len 3, we just get the salary
```

```python
salary_value = []

for i in range(0, df2.shape[0], 1):
    k = df2.salary[i].split(' ')
    if len(k) == 5:
        ## salary is average of salary min and max
        s_min = k[0].replace(',', '').replace('$', '')
        s_max = k[2].replace(',', '').replace('$', '')
        
        sv = (float(s_min) + float(s_max))/2.0
        salary_value.append(sv)
    else:
        s = k[0].replace(',', '').replace('$', '')
        sv = float(s)
        salary_value.append(sv)
        
df2['salary_value'] = salary_value
```

### convert all salary to by hours

```python
salary_by_hour=[]

## assume 1 year -- 52 work week, 5 day work and 8 hours per day
## 1 year = 52*5*8 = 2080 hr
## 1 week = 40 hr
## 1 day = 8 hr

for i in range(0,df2.shape[0],1):
    
    if df2.salary_duration[i] == 'year':
        sbh = df2.salary_value[i] / 2080.
        salary_by_hour.append(sbh)
        
    elif df2.salary_duration[i] == 'week':
        sbh = df2.salary_value[i] / 40.
        salary_by_hour.append(sbh)
        
    elif df2.salary_duration[i] == 'day':
        sbh = df2.salary_value[i] / 8.
        salary_by_hour.append(sbh)
        
    elif df2.salary_duration[i] == 'hour':
        sbh = df2.salary_value[i]
        salary_by_hour.append(sbh)

df2['salary_byhour'] = salary_by_hour

df2.head()
```

## Executive summary

1) Job Salary Trends

For factors determining salary job trend, the salary by the hour is first classify to either a higher then median or lower then median salary. I extracted the salary and convert it into salary by the hour for comparsion. An average was taken if there is a range of salary values. Words feature like company name, job description and job title we converted into numerical data matrix where the features are the number of times, a certain words was found in the dataset. Location feature was also used to look into the influence on classifying high or low salary. From the analysis, it is shown that location do not play a big role in determining the salary trend. This might be the case due to the fact that most higher then median salary data related job are found in NSW. The most important feature to determine salary trends are job title and job description. Words like analyst, data analytics are most common in high salary job. Also, it can be seen that most of the high paying data science related jobs comes from recruitment firm and IT consulting firm. 

2) Job Category Factors

For job category, I consider 2 target values. The first is either the job is data science related or not. The second is if the job is a senior job or junior job. These were extracted into binary form with 1 flagging data science job or senior job respectively. This time, only description features and job title features is consider. This problem is an imbalanced problem as most of these category are less than 10% of the dataset. Since most of the jobs comes from recruitment firm, it is hard to gauge what is the important feature for determining a job category. Also, the job summary scrap from au.indeed.com does not represent the full job description. A further work would be to enter each web link and get more data including job skill, requirement in job role and more job description.
